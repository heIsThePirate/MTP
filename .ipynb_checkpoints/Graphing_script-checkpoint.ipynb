```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset and dividing into corresponding numpy arrays
dataset = pd.read_csv("dft", sep = " ", header = None)

function = 'sin(2*PI*f_1*t/N) + sin(2*PI*f_2*t/N + phi)'

frequency = dataset.iloc[:, 0].values
function_value = dataset.iloc[:, 1].values
real_dft = dataset.iloc[:, 2].values
imag_dft = dataset.iloc[:, 3].values
psd = dataset.iloc[:, 4].values
```

```python
# Plotting the datasets
fig, ax = plt.subplots(3, sharex='col', figsize=(18, 16))

function = 'sin(2*PI*f_1*t/N) + sin(2*PI*f_2*t/N + phi)'

ax[0].plot(frequency, function_value)
ax[0].set(ylabel = function)

ax[1].plot(frequency, real_dft)
ax[1].set(ylabel = 'Real(dft)')

ax[2].plot(frequency, psd)
ax[2].set(xlabel = 'Frequency', ylabel = 'Power Spectrum')
```

```python
# Importing the 2-D data
data = pd.read_csv("dft_2D", sep = " ", header = None)

frequency = data.iloc[:, 0].values
function_value = data.iloc[:, 1].values
real_dft = data.iloc[:, 2].values
imag_dft = data.iloc[:, 3].values
psd = data.iloc[:, 4].values
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Plotting the 2-D sine wavefunction
fig = plt.figure(figsize=(12, 6))

ax = fig.add_subplot(1, 2,  1, projection='3d')

xvalues = np.arange(0, 100)
yvalues = np.arange(0, 100)

xgrid, ygrid = np.meshgrid(xvalues, yvalues)

zvalues = np.reshape(function_value, (100, 100))

surf = ax.plot_surface(xgrid, ygrid, zvalues, rstride=5, cstride=5, linewidth=0, cmap=cm.plasma)

ax = fig.add_subplot(1,2,2)

plt.contourf(xgrid, ygrid, zvalues, 30, cmap=cm.plasma)

fig.colorbar(surf, aspect=18)

plt.tight_layout()
```

```python
# Plotting the real part of DFT of the 2-D sine wavefunction
fig = plt.figure(figsize=(12, 6))

ax = fig.add_subplot(1, 2,  1, projection='3d')

xvalues = np.arange(0, 100)
yvalues = np.arange(0, 100)

xgrid, ygrid = np.meshgrid(xvalues, yvalues)

zvalues = np.reshape(real_dft, (100, 100))

surf = ax.plot_surface(xgrid, ygrid, zvalues, rstride=5, cstride=5, linewidth=0, cmap=cm.plasma)

ax = fig.add_subplot(1,2,2)

plt.contourf(xgrid, ygrid, zvalues, 30, cmap=cm.plasma)

fig.colorbar(surf, aspect=18)

plt.tight_layout()
```

```python
# Plotting the 'psd' of the 2-D sine wavefunction
fig = plt.figure(figsize=(12, 6))

ax = fig.add_subplot(1, 2,  1, projection='3d')

xvalues = np.arange(0, 100)
yvalues = np.arange(0, 100)

xgrid, ygrid = np.meshgrid(xvalues, yvalues)

zvalues = np.reshape(psd, (100, 100))

surf = ax.plot_surface(xgrid, ygrid, zvalues, rstride=5, cstride=5, linewidth=0, cmap=cm.plasma)

ax = fig.add_subplot(1,2,2)

plt.contourf(xgrid, ygrid, zvalues, 30, cmap=cm.plasma)

fig.colorbar(surf, aspect=18)

plt.tight_layout()
```
