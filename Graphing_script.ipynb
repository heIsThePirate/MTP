```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset and dividing into corresponding numpy arrays
dataset = pd.read_csv("dft", sep = " ", header = None)

function = 'sin(2*PI*f_1*t/N) + sin(2*PI*f_2*t/N + phi)'

frequency = dataset.iloc[:, 0].values
function_value = dataset.iloc[:, 1].values
real_dft = dataset.iloc[:, 2].values
imag_dft = dataset.iloc[:, 3].values
psd = dataset.iloc[:, 4].values
```

```python
# Plotting the datasets
fig, ax = plt.subplots(3, sharex='col', figsize=(18, 16))

function = 'sin(2*PI*f_1*t/N) + sin(2*PI*f_2*t/N + phi)'

ax[0].plot(frequency, function_value)
ax[0].set(ylabel = function)

ax[1].plot(frequency, real_dft)
ax[1].set(ylabel = 'Real(dft)')

ax[2].plot(frequency, psd)
ax[2].set(xlabel = 'Frequency', ylabel = 'Power Spectrum')
```
